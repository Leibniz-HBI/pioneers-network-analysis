```python
import pandas as pd
```

```python
edges_df = pd.read_csv('csv/graphdata_e.csv', header=None, names=['Source', 'Target', 'Source_Name', 'Target_Name'])
```

```python
edges_df
```

```python
sources_df = edges_df[['Source', 'Source_Name']]
```

```python
sources_df.drop_duplicates(ignore_index=True, inplace=True)
sources_df
```

```python
target_df = edges_df[['Target', 'Target_Name']]
target_df.drop_duplicates(ignore_index=True, inplace=True)
target_df
```

```python
sources_df.columns = ['Id', 'screen_name']
target_df.columns = ['Id', 'screen_name']
node_df = sources_df.append(target_df)
node_df
```

```python
edges_df = edges_df[['Source', 'Target']]
edges_df
```

```python
node_df.to_csv('csv/nodes.csv', index=False)
```
